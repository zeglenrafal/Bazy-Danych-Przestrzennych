```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itables import show, init_notebook_mode
import plotly.express as px
import itables.interactive as interactive
from itables import show
```

```python
init_notebook_mode(all_interactive=True)

file_path = "data.xlsx"
df = pd.read_excel(file_path, sheet_name=1)

show(df.head(100))
```

```python
df.info()
df.describe()
df.isna().sum()
```

```python
# Przykład grupowania po kraju
country_group = df.groupby('WHO Country Name')[['PM2.5 (μg/m3)', 'PM10 (μg/m3)', 'NO2 (μg/m3)']].mean()
country_group.sort_values('PM2.5 (μg/m3)', ascending=False).head(10)
```

```python
# Wykresy rozkładu
sns.boxplot(data=df, y='PM2.5 (μg/m3)')
plt.title('Rozkład PM2.5')
plt.show()
```

```python
country_group.plot(kind="bar", figsize=(12, 6), title="Średnie stężenie zanieczyszczeń wg regionów")
plt.ylabel("μg/m3")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
pollutants = {
    "PM2.5 (μg/m3)": "Top 10 countries by PM2.5 concentration",
    "PM10 (μg/m3)": "Top 10 countries by PM10 concentration",
    "NO2 (μg/m3)": "Top 10 countries by NO2 concentration"
}

# Dla każdego rodzaju zanieczyszczenia wybierz top 10 miast
for col, title in pollutants.items():
    top_cities = (
        df[["WHO Country Name", col]]
        .dropna()
        .groupby("WHO Country Name")
        .mean()
        .sort_values(by=col, ascending=False)
        .head(10)
        .reset_index()
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_cities, x=col, y="WHO Country Name", palette="magma")
    plt.title(title)
    plt.xlabel(f"{col}")
    plt.ylabel("Country")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
pollutants_regions = {
    "PM2.5 (μg/m3)": "Top 10 regions by PM2.5 concentration",
    "PM10 (μg/m3)": "Top 10 regions by PM10 concentration",
    "NO2 (μg/m3)": "Top 10 regions by NO2 concentration"
}

for col, title in pollutants_regions.items():
    top_cities = (
        df[["WHO Region", col]]
        .dropna()
        .groupby("WHO Region")
        .mean()
        .sort_values(by=col, ascending=False)
        .head(10)
        .reset_index()
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_cities, x=col, y="WHO Region", palette="magma")
    plt.title(title)
    plt.xlabel(f"{col}")
    plt.ylabel("Country")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
pollutants_cities = {
    "PM2.5 (μg/m3)": "Top 30 cities by PM2.5 concentration",
    "PM10 (μg/m3)": "Top 30 cities by PM10 concentration",
    "NO2 (μg/m3)": "Top 30 cities by NO2 concentration"
}

for col, title in pollutants_cities.items():
    top_cities = (
        df[["City or Locality", col]]
        .dropna()
        .groupby("City or Locality")
        .mean()
        .sort_values(by=col, ascending=False)
        .head(30)
        .reset_index()
    )

    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_cities, x=col, y="City or Locality", palette="magma")
    plt.title(title)
    plt.xlabel(f"{col}")
    plt.ylabel("Country")
    plt.grid(axis="x", linestyle="--", alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import folium
```

```python


top_pm25 = (
    df[['City or Locality', 'WHO Country Name', 'PM2.5 (μg/m3)']]
    .dropna()
    .groupby(['City or Locality', 'WHO Country Name'])
    .mean()
    .sort_values(by='PM2.5 (μg/m3)', ascending=False)
    .head(100)
    .reset_index()
)


geolocator = Nominatim(user_agent="air_pollution_project")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)


def get_coordinates(row):
    location = geocode(f"{row['City or Locality']}, {row['WHO Country Name']}")
    if location:
        return pd.Series([location.latitude, location.longitude])
    else:
        return pd.Series([None, None])


top_pm25[['Latitude', 'Longitude']] = top_pm25.apply(get_coordinates, axis=1)


map_pm25 = folium.Map(location=[20, 0], zoom_start=2)


for _, row in top_pm25.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=7,
            popup=f"{row['City or Locality']}, {row['WHO Country Name']}<br>PM2.5: {row['PM2.5 (μg/m3)']:.1f} μg/m3",
            color='red',
            fill=True,
            fill_color='red'
        ).add_to(map_pm25)


map_pm25
```

```python
# Grupowanie danych według krajów i obliczenie średniego stężenia PM2.5
country_pm25 = (
    df[['WHO Country Name', 'PM2.5 (μg/m3)']]
    .dropna()
    .groupby('WHO Country Name')
    .mean()
    .reset_index()
)

# Tworzenie mapy z użyciem plotly
fig = px.choropleth(
    country_pm25,
    locations="WHO Country Name",
    locationmode="country names",
    color="PM2.5 (μg/m3)",
    hover_name="WHO Country Name",
    color_continuous_scale="magma",
    title="Średnie stężenie PM2.5 wg kraju (μg/m³)"
)

fig.update_layout(geo=dict(showframe=False, showcoastlines=False))
fig.show()
```
